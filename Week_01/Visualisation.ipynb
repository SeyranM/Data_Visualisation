```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('food_consumption.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
names = set(df["food_category"])
colors = ['red', 'blue', 'green', 'black', 'purple', 'yellow', 'maroon', 'olive', 'aqua', 'fuchsia', 'lime']
grouped_by_fc = df.groupby(["food_category"])
fig, axes = plt.subplots(3, 4, figsize=(15, 15))
n_iter = 0
for name, data in grouped_by_fc:
    axes.flatten()[n_iter].set_title(name)
    axes.flatten()[n_iter].scatter(data['consumption'],data['co2_emmission'], c = colors[n_iter]);
    n_iter += 1
```
